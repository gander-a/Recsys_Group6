Parse attributes containing tab-separated lists into lists. Transform raw timestamps into human-readable timestamps.

```python
df['text_tokens'] = df['text_tokens'].str.split('\t')

def to_hex_list(x):
    output = str(x).split('\t')
#     output = [int(val, 16) for val in str(x).split('\t')] 
    return output

cols_to_process = ['hashtags', 'present_media', 'present_links', 'present_domains']

for col in cols_to_process:  
    df[col] = df[col].apply(lambda x: to_hex_list(x) if isinstance(x, str)  else x)


    
cols_to_process = ['tweet_timestamp', 'engaging_user_account_creation', 'reply_timestamp', 'retweet_timestamp', 'retweet_with_comment_timestamp', 'like_timestamp']

for col in cols_to_process:  
    df[col] = df[col].apply(lambda x: pd.Timestamp(x, unit='s'))
```

```python
def has_timestamp(x):
  return int(not pd.isnull(x))

def create_labels(col):
  copy_test = X_test.copy()
  copy_test['label'] = copy_test.apply(lambda row: has_timestamp(row[col]), axis = 1)
  return copy_test[['engaging_user_id', 'tweet_id', 'label']]

def write_groundtruth(engagement):
  gt = create_labels(engagement + '_timestamp')
  gt.to_csv('gt_' + engagement + '.csv', index = False)
    
def read_predictions(file):
  return pd.read_csv(file, names = ['user_id', 'tweet_id', 'label'], header = 0)

write_groundtruth('reply')
write_groundtruth('retweet')
write_groundtruth('retweet_with_comment')
write_groundtruth('like')
```

## T3 - Create a Ratings Matrix

```python
from scipy import sparse as sp
```

```python
unique_user_ids = df['engaging_user_id'].append(df['engaged_with_user_id']).unique()
unique_user_ids.sort()

unique_tweet_ids = df['tweet_id'].unique()
unique_tweet_ids.sort()

m = len(unique_user_ids)
n = len(unique_tweet_ids)

userId_to_userIDX = dict(zip(unique_user_ids, range(m)))
userIDX_to_userId = dict(zip(range(m), unique_user_ids))

tweetId_to_tweetIDX = dict(zip(unique_tweet_ids, range(n)))
tweetIDX_to_tweetId = dict(zip(range(n), unique_tweet_ids))
```

```python
cols = ['tweet_id', 'engaging_user_id', 'reply_timestamp', 'retweet_timestamp',
       'retweet_with_comment_timestamp', 'like_timestamp']

r_df = pd.concat([df['engaging_user_id'].map(userId_to_userIDX),
                  df['tweet_id'].map(tweetId_to_tweetIDX),
                  df['reply_timestamp'].notnull(),
                  df['retweet_timestamp'].notnull(),
                  df['retweet_with_comment_timestamp'].notnull(),
                  df['like_timestamp'].notnull()], axis = 1)

r_df.columns = ['user', 'tweet', 'reply', 'retweet', 'retweet_wc', 'like']
r_df.sort_values(['user', 'tweet'], inplace = True)

r_df.head(n = 20)
```

```python
R_reply = sp.csr_matrix((r_df.reply.astype('int8') * 2 -1, (r_df.user, r_df.tweet)), 
                        shape=(m, n))

R_retweet = sp.csr_matrix((r_df.retweet.astype('int8') * 2 -1, (r_df.user, r_df.tweet)), 
                          shape=(m, n))

R_retweet_wc = sp.csr_matrix((r_df.retweet_wc.astype('int8') * 2 -1, (r_df.user, r_df.tweet)), 
                             shape=(m, n))

R_like = sp.csr_matrix((r_df.like.astype('int8') * 2 -1, (r_df.user, r_df.tweet)), 
                       shape=(m, n))

display(R_reply.shape)
display(R_retweet.shape)
display(R_retweet_wc.shape)
display(R_like.shape)
```

## User-user collaborative filtering

The following code computes the average rating of each user.

```python
user_sums = R_like.sum(axis=1).A1 ## matrix converted to 1-D array via .A1
user_avgs = user_sums / n # devide through total number of tweets to get percentage of interaction
print("R_like", R_like)
print("user_sums[4]", user_sums[4])
print("user_avgs[4]", user_avgs[4])
```

### User-User Similarity (Not needed - Just for Debugging)

The following function computes the mean-centered cosine similarity between two users.

```python
from scipy.sparse.linalg import norm

def compute_pairwise_user_similarity(u_id, v_id):
    u = R_like[u_id,:].copy()
    v = R_like[v_id,:].copy()
    
    numerator_matrix = u.dot(v.T)
    numerator = numerator_matrix[0,0]

    denominator = norm(u) * norm(v)
    
    if denominator == 0:
        similarity = 0.;
    else:
        similarity = numerator/denominator
    
    return similarity
```

### User to all Users Similarities

The following function computes the cosine similarities of a given user to all other users.

```python
import sklearn.preprocessing as pp

def compute_user_similarities(u_id):
    uU = np.empty((m,))

    R_avg = np.asmatrix(user_avgs).transpose()
    R_avg = sp.diags(user_avgs)

    R_avg = R_avg * R_like.astype(bool).astype(int)
    R_like_copy = R_like - R_avg
    pp.normalize(R_like_copy, copy=False)
    u = R_like_copy[u_id,:]
    uU = R_like_copy.dot(u.T)
    
    return uU.toarray().flatten()
```

### Create User Neighborhood

The following function creates the user neighborhood of a given user. It takes as input, the target user u_id and the target itemi_id, and uses additional parameters, the size k of the neighborhood.

```python
R_like_dok = R_like.todok()

## default values
k = 5

def create_user_neighborhood(u_id, i_id):
    nh = {} ## the neighborhood dict with (user id: similarity) entries
    ## nh should not contain u_id and only include users that have interacted with i_id; there should be at most k neighbors
    uU = compute_user_similarities(u_id)
    uU_copy = uU.copy() ## so that we can modify it, but also keep the original

    uU_sort = np.argsort(uU_copy)[::-1]
        
    i = 0
    while len(nh) < k and i < len(uU_sort):
        if uU_sort[i] != u_id and (uU_sort[i], i_id) in R_like_dok:
            nh[uU_sort[i]] = uU[uU_sort[i]].item(0)
        i += 1

    return nh
```

### Predict a Probability (internal_ids)

```python
def predict_rating_internal(u_id, i_id):

    if (u_id, i_id) in R_like_dok:
        print("user", u_id, "has interacted on item", i_id, "with", R_like_dok[u_id, i_id])
    else:
        print("user", u_id, "has not interacted with item", i_id)
        print("k:", k)


    nh = create_user_neighborhood(u_id, i_id)

    neighborhood_weighted_avg = 0.
    numerator = 0.
    denominator = 0.

    for v in nh:
        numerator += nh[v] * R_like[v,i_id]

        denominator += np.absolute(nh[v])


    if denominator == 0:
        neighborhood_weighted_avg = 0.;
    else:
        neighborhood_weighted_avg = numerator/denominator


    prediction = neighborhood_weighted_avg
    print(f'prediction {prediction:.4f} (user_avg {user_avgs[u_id]:.4f})')
    print("")
    return prediction
    
```

### Predict a Probability (External IDs)

```python
def predict_rating(engaging_user_id, tweet_id):
    print("user", engaging_user_id, "has internal id ", userId_to_userIDX[engaging_user_id])
    print("tweet", tweet_id, "has internal id ", tweetId_to_tweetIDX[tweet_id])
    return predict_rating_internal(userId_to_userIDX[engaging_user_id],tweetId_to_tweetIDX[tweet_id])
```
